## 2. Import Classes

```python
from sezim_emolex import SezimEmoLex, SezimAnalyzer
```

## 3. Analyze a Single Word

Let's analyze the word `"ашу"` (anger).

```python
word = "ашу"
lex = SezimEmoLex(word)

print(f"Word: {lex.word}
")
print(f"Found in lexicon: {lex.found}
")
print(f"Sentiment: {lex.get_sentiment_label()} ({lex.get_sentiment()})
")
print(f"Emotions: {lex.get_emotions()}
")
print(f"Does it have 'anger'? {lex.has_emotion('anger')}")
```

## 4. Analyze a Full Text

Now, let's analyze a complete sentence.

```python
analyzer = SezimAnalyzer()

text = "мен қатты ашуландым бірақ соңында бәрі жақсы болды"
result = analyzer.analyze(text)

import json
print(json.dumps(result, indent=2, ensure_ascii=False))
```

### Exploring the Results

```python
print(f"Text: '{result['text']}'
")
print(f"Words found: {result['words_found']}/{result['total_words']}
")
print(f"Overall Sentiment: {result['sentiment_label']} (Score: {result['sentiment_score']})
")
print(f"Emotion Scores: {result['emotions']}")
```

## 5. Working with the Lexicon Directly

You can also access the underlying lexicon data.

```python
# The lexicon is loaded as a class variable, so we can access it
lexicon_df = SezimEmoLex._get_lexicon()

print(f"Total words in lexicon: {len(lexicon_df)}
")
print("First 5 rows:")
lexicon_df.head()
```

### Find all words with the 'joy' emotion

```python
joy_words = lexicon_df[lexicon_df['joy'] == 1]

print(f"Found {len(joy_words)} words with 'joy'.
")
print("Top 10 words with joy:")
print(joy_words['word'].head(10).tolist())
```
